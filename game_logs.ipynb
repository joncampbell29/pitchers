```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("stats.csv")
df.rename(columns={'last_name, first_name': 'name'}, inplace=True)
df[['last_name', 'first_name']] = df['name'].str.split(' ', n=1, expand=True)
```

```python
def get_pitcher_gameslogs(first,last,id):
  full = "-".join([first, last, str(id)])
  url = f"https://baseballsavant.mlb.com/savant-player/{full}"
  payload = {
      'stats': 'gamelogs-r-pitching-mlb',
      'season': '2024'
      }
  resp = requests.get(url, params=payload)
  print(resp.url)
  soup = BeautifulSoup(resp.text, 'html.parser')
  table = soup.find(id="gamelogs-mlb").find_all('table')[0]
  df = pd.read_html(str(table))[0]
  return df
```

```python
test = dict(df.iloc[126])
print(test['name'])
```

```python
t_first, t_last, t_id = test['first_name'], test['last_name'], test['player_id']
```

```python
t = get_pitcher_gameslogs(t_first, t_last, t_id)
```
